```python
plt.figure(figsize=(10, 6))
plt.bar(['Ideal Sıcaklık (°C)', 'Maksimum Sıcaklık (°C)', 'Minimum Sıcaklık (°C)'], [16, 40, -4], color=['green', 'darkgreen', 'brown'])
plt.title('Avokado Ağacı İçin İdeal Sıcaklık Aralığı')
plt.ylabel('Sıcaklık (°C)')
plt.show()
```

```python
ozellikler = ['Yıllık Yağış (mm)', 'Maksimum Sıcaklık (°C)', 'Minimum Sıcaklık (°C)']
veriler = [800, 1700, 40, -4]


plt.figure(figsize=(8, 6))
bars = plt.bar(ozellikler, veriler[:3], color=['blue', 'red', 'green'])

for bar, value in zip(bars, veriler[:3]):
    plt.text(bar.get_x() + bar.get_width()/2, value + 20, str(value), ha='center', va='bottom', fontsize=12)


plt.title('Avokado Ağacı İçin İklim Verileri')
plt.ylabel('Değer')
plt.ylim(-50, 1800)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
ozellikler = ['pH Değeri', 'Drenaj', 'Tuzluluk', 'Toprak Türü']
aciklamalar = ['5 - 7 (hafif asidik ila nötr)',
               'İyi drene olan topraklar tercih edilmelidir.',
               'Düşük olmalıdır.',
               'Tınlı, kumlulu veya gevşek yapılı topraklar tercih edilir.']

plt.figure(figsize=(15, 8))
bars = plt.barh(ozellikler, range(len(ozellikler)), color='skyblue', alpha=0.7)


for i, txt in enumerate(aciklamalar):
    plt.text(0.5, i, txt, va='center', ha='left', wrap=True, fontsize=12)


plt.xlim(0, 1)
plt.yticks(range(len(ozellikler)), ozellikler, fontsize=12)
plt.xlabel('Açıklama', fontsize=12)
plt.title('Avokado Ağacı İçin Toprak Analizi', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
